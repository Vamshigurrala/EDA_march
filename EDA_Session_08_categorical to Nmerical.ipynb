**One hot encoder**
- one hot means if one is ON another is OFF
- ON represents with 1, and OFF represents with 0
- For example case status has two unique labels
    - Denied
    - Certified
- if we apply one hot encoder on case status column it will create two extra columns
    - case status denied
    - case status certified

|case_status|case_status_denied|case_status_certified|
|--------------|----|----|
|Denied|1|0|
|Certified|0|1|

**Advantages**

- In machine learning we have one important property is **independent of variables**
- machine learning models expects variables are independent each other
- one hot encoder creating new columns, the new columns are **independent each other**
- independent variables means both variable have **90 degrees phase shift**
- 90 degrees phase shift also called **perpendicular each other**
- perpendicular means **orthogonal each other**

**Disadvantages**
- one hot encoder creates new columns equal to number of unique labels
- if number of unique labels increase new columns also increase
- the more columns lead to more data
- if data is more then complexity will increase
- if complexity increase more computation power required
- time complexity also increases
- the entire process named as **curse of dimensionality**
